```python
features = ["Age", "Income", "Total_Spending", "NumWebPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Recency"]
```

```python
X = df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2, 10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2, 10), wcss, marker = "o")
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 6)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
df.head()
```

```python
cluster_summary = df.groupby("Cluster")[features].mean()
```

```python
cluster_summary
```

```python
df["Cluster"].value_counts()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_data = pca.fit_transform(X_scaled)
df["PCA1"] = pca_data[:, 0]
df["PCA2"] = pca_data[:, 1]
```

### Interpreting Customer Clusters

Let's visualize the average values of the features for each cluster to understand their characteristics.

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 15))
axes = axes.flatten()

for i, col in enumerate(cluster_summary.columns):
    if i < len(axes):
        sns.barplot(
            x=cluster_summary.index,
            y=col,
            data=cluster_summary,
            hue=cluster_summary.index,   # ✅ Add this
            palette='viridis',
            legend=False,                # ✅ Remove legend
            ax=axes[i]
        )
        axes[i].set_title(f'Average {col} by Cluster')
        axes[i].set_xlabel('Cluster')
        axes[i].set_ylabel(f'Average {col}')

plt.tight_layout()
plt.show()
```

```python
import joblib

# Save the KMeans model
joblib.dump(kmeans, 'kmeans_model.pkl')

# Save the StandardScaler
joblib.dump(scaler, 'scaler.pkl')
```

These plots help us understand the unique profile of each customer cluster:

*   **Cluster 0:** High income, high total spending, older, low web visits, high store purchases, high recency.
*   **Cluster 1:** Lower income, low total spending, medium age, high web visits, low store purchases, low recency.
*   **Cluster 2:** Medium-high income, medium-high total spending, older, high web purchases, high web visits, medium recency.
*   **Cluster 3:** Highest income, highest total spending, youngest, low web visits, high store purchases, medium recency.
*   **Cluster 4:** Low income, low total spending, medium age, low web purchases, low store purchases, high recency.
*   **Cluster 5:** High income, high total spending, older, low recency, very high store purchases.

Based on these profiles, we can give meaningful names to each cluster, such as:
*   **Cluster 0:** Senior High-Spending Traditional Customers
*   **Cluster 1:** Low-Income Browsers (Low Converters)
*   **Cluster 2:** Active Multi-Channel Shoppers
*   **Cluster 3:** Young High-Value Premium Customers
*   **Cluster 4:** At-Risk Low-Value Customers
*   **Cluster 5:** Loyal High-Value Store Customers
